```python
import SGD
import computeFunctions as cf
import utils
import math
import numpy as np
import seaborn as sns
import pandas as pd
```

# Define the parameters for the optimization

```python
alpha_zero = 1
c1 = 0.1
beta = 0.5
epsilon = 1e-5
```

# Study the behaviour with different matrices of given size and density

```python
#perform 10 tests usign Rayleigh quotient and backtracking
lists_gradients = []
lists_alphas = []
matrices = []
starting_points = []
for i in range(10):
    A = utils.randomMatrix(100, 100, 1)
    matrices.append(A)
    x = np.random.rand(100)
    starting_points.append(x)
    res = SGD.SGD(cf.computeRayleigh, cf.computeGradientRayleigh, x, alpha_zero, beta, c1, epsilon, A, 1)
    lists_gradients.append(res[0])
    lists_alphas.append(res[1])
```

```python
lists_gradients = utils.ResultsMatrix(lists_gradients)
```

```python
gradientsDF = pd.DataFrame(lists_gradients).melt()
```

```python
chart = sns.lineplot(data=gradientsDF, y='value', x='variable')
chart.set(yscale='log')
```

```python
lists_alphas = utils.ResultsMatrix(lists_alphas)
```

```python
alphasDF = pd.DataFrame(lists_alphas).melt()
```

```python
chart = sns.lineplot(data=alphasDF, y='value', x='variable')
chart.set(yscale='log')
```

# Test with the  new function: same matrices, starting point and parameters as above

```python
#test using the new function
lists_gradients = []
lists_alphas = []
for i in range(len(matrices)):
    x = starting_points[i]
    A = matrices[i]
    Q = np.dot(A.T, A)
    res = SGD.SGD(cf.computeNewFunction, cf.computeGradientNewFunction, x, alpha_zero, beta, c1, epsilon, Q, 1)
    lists_gradients.append(res[0])
    lists_alphas.append(res[1])
```

```python
lists_gradients = utils.ResultsMatrix(lists_gradients)
```

```python
gradientsDF = pd.DataFrame(lists_gradients).melt()
```

```python
chart = sns.lineplot(data=gradientsDF, y='value', x='variable')
chart.set(yscale='log')
```

```python
lists_alphas = utils.ResultsMatrix(lists_alphas)
```

```python
alphasDF = pd.DataFrame(lists_alphas).melt()
```

```python
chart = sns.lineplot(data=alphasDF, y='value', x='variable')
chart.set(yscale='log')
```
